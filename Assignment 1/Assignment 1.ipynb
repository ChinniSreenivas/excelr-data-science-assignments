```python
# 1.prime number program

num = int(input( "enter a number  "))
flag = 1

for i in range(2, num//2) :
      rem = num % i
      if rem == 0 :
          flag = 0
          break
      else :
          flag = 1


if flag == 1 and num >=2:
   print(" its prime number   ")
else :
   print(" its not prime number ");
```

```python
# 2. Product of two random numbers
import random
```

```python
# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user to enter the product
print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Your answer: "))

# Check if the user's answer is correct
correct_answer = num1 * num2
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {correct_answer}.")
```

```python
# Exercise 3: Squares of Even Numbers from 100 to 200
for num in range(100, 201):
    if num % 2 == 0:  # Check if the number is even
        square = num ** 2
        print(f"The square of {num} is {square}")
```

```python
4. # Hospital Billing System

# Given details
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge_per_test = 300
medicine_charges = 2400

# Patient's usage
days_stayed = 4
number_of_lab_tests = 3

# Calculate the total bill before discount
room_charges = room_charge_per_day * days_stayed
lab_test_charges = lab_test_charge_per_test * number_of_lab_tests

total_before_discount = room_charges + doctor_fee + lab_test_charges + medicine_charges

# Apply discount if applicable
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

final_bill = total_before_discount - discount

# Print the details
print("Hospital Billing Summary")
print("------------------------")
print(f"Room Charges for {days_stayed} days: ₹{room_charges}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges for {number_of_lab_tests} tests: ₹{lab_test_charges}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Total before discount: ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
```
